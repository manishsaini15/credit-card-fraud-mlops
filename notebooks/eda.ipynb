```python
# Import pandas for data analysis
import pandas as pd

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Read original dataset
df = pd.read_csv("../artifacts/creditcard.csv")

# Display first 5 rows
df.head()
```

```python
# Print shape of dataset
print("Dataset shape:", df.shape)

# Count fraud and legitimate transactions
print(df["Class"].value_counts())

# Show class ratio in percentage
print(df["Class"].value_counts(normalize=True) * 100)
```

```python
# Display statistical summary of Amount column
df["Amount"].describe()
```

```python
# Separate legitimate transactions
legit = df[df["Class"] == 0]

# Separate fraud transactions
fraud = df[df["Class"] == 1]

# Compare average amount
print("Legit mean amount:", legit["Amount"].mean())
print("Fraud mean amount:", fraud["Amount"].mean())
```

```python
# Show summary statistics of Time column
df["Time"].describe()
```
